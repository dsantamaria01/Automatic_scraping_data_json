```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains 
from selenium.webdriver.common.keys import Keys # 
import os
 

page = 'https://ofdollarsanddata.com/investment-quotes/'
driver = webdriver.Edge()
driver.get(page)
driver.maximize_window()

time.sleep(5)

todos_p = driver.find_elements(By.TAG_NAME, 'p')

quotes = []
for linea in todos_p:
    if linea.text.find('“') == 1 or linea.text.find('“') == 0:
        lon1 = len(linea.text)
        quotes.append(linea.text[1:lon1-1])
    if  linea.text.find('-') == 0:
        lon2 = linea.text.find(',')
        if lon2 == -1:
            lon2 = len(linea.text)
        quotes.append(linea.text[1:lon2])

pares=[]
impares=[]
for num in range(len(quotes)):
    if (num % 2) == 0:  
        pares.append(quotes[num])
    else:  
        impares.append(quotes[num])
        
from openpyxl import Workbook

wb = Workbook()
wb.create_sheet("UPWORK")
wb.create_sheet("Sheet_two")

wb.sheetnames
['UPWORK', 'Sheet_one', 'Sheet_two']

ws1 = wb['UPWORK']

ws1['A1'] = 'QUOTES'
ws1['B1'] = 'NAMES'

row_start = 2  #start below the header row 2
col_start = 1  #starts from column B
for i in range(0,len(pares)):
    ws1.cell(row_start+i, col_start).value = pares[i]
    ws1.cell(row_start+i, col_start+1).value = impares[i]
    
    


wb.save('UPWORK.xlsx')
```
